Jeu de données 1 : YEAR, (Kw)

Régression linéaire

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le jeu de données
url = "https://raw.githubusercontent.com/rashida048/Datasets/9fefabe05e3b685bf741a25f506654adeb71ff69/cars.csv"
df = pd.read_csv(url)
# Sélectionner les colonnes year et price
X = df['YEAR']
y = df['(kW)']

# Calculer la régression linéaire
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X.values.reshape(-1, 1), y)

# Tracer la série temporelle et la tendance
plt.figure(figsize=(12, 6))
plt.plot(X, y, label='Prix')
plt.plot(X, model.predict(X.values.reshape(-1, 1)), label='Tendance (régression linéaire)')
plt.xlabel('Année')
plt.ylabel('Prix')
plt.title('Évolution des prix des voitures')
plt.legend()
plt.show()
```

Moyenne mobile (fenêtre de 3 périodes) :

```python
# Calculer la moyenne mobile
df['Price_MA'] = df['(kW)'].rolling(window=3).mean()

# Tracer la série temporelle et la tendance
plt.figure(figsize=(12, 6))
plt.plot(df['YEAR'], df['(kW)'], label='(kW)')
plt.plot(df['YEAR'], df['Price_MA'], label='Tendance (moyenne mobile)')
plt.xlabel('Année')
plt.ylabel('(kW)')
plt.title('Évolution des (kW) des voitures')
plt.legend()
plt.show()
```

Lissage exponentiel (alpha = 0.2) :

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Calculer le lissage exponentiel
df['(kW)_ES'] = df['(kW)'].copy()
model = SimpleExpSmoothing(df['(kW)_ES']).fit(smoothing_level=0.2)
df['(kW)_ES'] = model.fittedvalues

# Tracer la série temporelle et la tendance
plt.figure(figsize=(12, 6))
plt.plot(df['YEAR'], df['(kW)'], label='(kW)')
plt.plot(df['YEAR'], df['(kW)_ES'], label='Tendance (lissage exponentiel)')
plt.xlabel('Année')
plt.ylabel('(kW)')
plt.title('Évolution des (kW) des voitures')
plt.legend()
plt.show()
```

Jeu de données 2 : year, mileage

Régression linéaire :

```python
# Charger le jeu de données

url = "https://raw.githubusercontent.com/rashida048/Datasets/9fefabe05e3b685bf741a25f506654adeb71ff69/cars.csv"
df = pd.read_csv(url)
# Sélectionner les colonnes year et CITY (kWh/100 km)
X = df['YEAR']
y = df['CITY (kWh/100 km)']

# Calculer la régression linéaire
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X.values.reshape(-1, 1), y)

# Tracer la série temporelle et la tendance
plt.figure(figsize=(12, 6))
plt.plot(X, y, label='CITY (kWh/100 km)')
plt.plot(X, model.predict(X.values.reshape(-1, 1)), label='Tendance (régression linéaire)')
plt.xlabel('Année')
plt.ylabel('CITY (kWh/100 km)')
plt.title('Évolution du CITY (kWh/100 km)	 des voitures')
plt.legend()
plt.show()
```

```python
Moyenne mobile (fenêtre de 3 périodes) :
```

```python
# Calculer la moyenne mobile
df['CITY (kWh/100 km)_MA'] = df['CITY (kWh/100 km)'].rolling(window=3).mean()

# Tracer la série temporelle et la tendance
plt.figure(figsize=(12, 6))
plt.plot(df['YEAR'], df['CITY (kWh/100 km)'], label='Kilométrage')
plt.plot(df['YEAR'], df['CITY (kWh/100 km)_MA'], label='Tendance (moyenne mobile)')
plt.xlabel('Année')
plt.ylabel('CITY (kWh/100 km)	')
plt.title('Évolution du CITY (kWh/100 km)	 des voitures')
plt.legend()
plt.show()
```

Lissage exponentiel (alpha = 0.2) :

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Calculer le lissage exponentiel
df['CITY (kWh/100 km)_ES'] = df['CITY (kWh/100 km)'].copy()
model = SimpleExpSmoothing(df['CITY (kWh/100 km)_ES']).fit(smoothing_level=0.2)
df['CITY (kWh/100 km)_ES'] = model.fittedvalues

# Tracer la série temporelle et la tendance
plt.figure(figsize=(12, 6))
plt.plot(df['YEAR'], df['CITY (kWh/100 km)'], label='CITY (kWh/100 km)')
plt.plot(df['YEAR'], df['CITY (kWh/100 km)_ES'], label='Tendance (lissage exponentiel)')
plt.xlabel('Année')
plt.ylabel('CITY (kWh/100 km)')
plt.title('Évolution du CITY (kWh/100 km) des voitures')
plt.legend()
plt.show()
```
